## Models Training and Evaluation

```python
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from joblib import dump
import os
```

```python
direct = '/home/jovyan/E2E_ML_Projects_Course/Project/Data'
```

```python
# 1. Load preprocessed data
X_train = pd.read_csv(f'{direct}/X_train_preprocessed.csv')
X_test = pd.read_csv(f'{direct}/X_test_preprocessed.csv')
y_train = pd.read_csv(f'{direct}/y_train.csv').squeeze()  
y_test = pd.read_csv(f'{direct}/y_test.csv').squeeze()
```

```python
# 2. Create directory to save models 
os.makedirs('saved_models', exist_ok=True)
```

```python
# 3. Define and train models
```

```python
print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of y_train: {y_train.shape}")
```

```python
models = {
    'LinearRegression': LinearRegression(),
    'RandomForest': RandomForestRegressor(n_estimators=100, random_state=42),
    'GradientBoosting': GradientBoostingRegressor(n_estimators=100, random_state=42)
}
```

```python
results = {}
for name, model in models.items():
    print(f"\nTraining {name}...")
    
    # Train model
    model.fit(X_train, y_train)
    
    # Evaluate
    y_pred = model.predict(X_test)
    results[name] = {
        'MSE': mean_squared_error(y_test, y_pred),
        'R2': r2_score(y_test, y_pred)
    }
    
    # Save model
    dump(model, f'saved_models/{name}.joblib')
    print(f"{name} saved to saved_models/{name}.joblib")
```

```python
# 4. results
print("\nModel Performance:")
print(pd.DataFrame(results).T)

# 5. Save evaluation results
pd.DataFrame(results).T.to_csv('saved_models/model_performance.csv')
print("\nEvaluation results saved to saved_models/model_performance.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from joblib import load
import pandas as pd
import numpy as np
```

```python
m_home = '/home/jovyan/E2E_ML_Projects_Course/Project/saved_models'
```

```python
# 1. Load test data and models
X_test = pd.read_csv(f'{direct}/X_test_preprocessed.csv')
y_test = pd.read_csv(f'{direct}/y_test.csv').squeeze()

models = {
    'Linear Regression': load(f'{m_home}/LinearRegression.joblib'),
    'Random Forest': load(f'{m_home}/RandomForest.joblib'),
    'Gradient Boosting': load(f'{m_home}/GradientBoosting.joblib')
}
```

```python
# 2. Set up plotting
plt.figure(figsize=(15, 10))
sns.set_style('whitegrid')

# 3. Generate residual plots for each model
for i, (name, model) in enumerate(models.items(), 1):
    # Get predictions and residuals
    y_pred = model.predict(X_test)
    residuals = y_test - y_pred
    
    # Create subplots
    plt.subplot(2, 3, i)
    sns.scatterplot(x=y_pred, y=residuals, alpha=0.5)
    plt.axhline(y=0, color='r', linestyle='--')
    plt.title(f'{name} Residuals vs Predicted')
    plt.xlabel('Predicted Values')
    plt.ylabel('Residuals')
    
    plt.subplot(2, 3, i+3)
    sns.histplot(residuals, kde=True, bins=30)
    plt.title(f'{name} Residual Distribution')
    plt.xlabel('Residuals')

plt.tight_layout()
plt.savefig(f'{m_home}/residual_plots.png', dpi=300)
plt.show()
```

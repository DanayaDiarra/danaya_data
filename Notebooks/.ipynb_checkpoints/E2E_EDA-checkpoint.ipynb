```python
# SCATTER PLOT OF ODOMETER IMPACT ON CAR PRICE 
plt.figure(figsize=(10, 6))
sns.histplot(x='odometer', y='sellingprice', data=data)
plt.title('Relationship between Odometer and Selling Price')
plt.tight_layout()
plt.show()
```

```python
#AVERAGE CCAR PRICE BASED ON CONDITION
avg_price_condition = data.groupby('condition')['sellingprice'].agg(['mean', 'count']).round(2)
sorted_avg_price_condition = sorted(avg_price_condition)

plt.figure(figsize= (12,8))
avg_price_condition['mean'].plot(kind='barh', color='grey')
plt.title('Average Selling Price by Vehicle Condition')
plt.xlabel('Average Price')
plt.ylabel('Condition')
plt.xticks(rotation=0)
plt.tight_layout()
print("\
Average Price by Condition:")
print(avg_price_condition)
plt.show()
```

```python
unique_years = data['year'].unique()
sorted_U_years = sorted(unique_years)
sorted_U_years
```

```python
plt.figure(figsize = (9, 8))
sns.histplot(data['year'], bins = sorted_U_years, kde = True)
plt.title('Vehicules Sales by Year')
plt.xlabel('Year')
plt.ylabel('Number of vehicules Sold')
# plt.grid()
plt.show()
```

```python
# CALCULATING CARS' AGES RANGE
current_year = 2025
data['car_age'] = current_year - data['year']
print(data[['year', 'car_age']])


plt.figure(figsize = (9, 8))
sns.histplot(data['car_age'], bins = 40, kde = True)
plt.title('Distribution of Cars Ages')
plt.xlabel('Age of Car (years)')
plt.ylabel('Number of vehicules')
plt.show()
```

```python
# MAKE/BRAND
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(25, 17))
plt.figure(figsize=(26, 25))

ax1.set_title('Number of cars by make(brand)')
sns.countplot(data['make'], order = data['make'].value_counts().index, ax=ax1)

ax2.set_title('Selling price by make(brand)')
sns.boxplot(x='sellingprice', y='make', data=data, order=data['make'].value_counts().index, ax=ax2)
```

```python
# Box plot of prices by brand/make (top 10 makes by frequency)
plt.figure(figsize=(12, 6))
top_makes = data['make'].value_counts().nlargest(10).index
df_top_makes = data[data['make'].isin(top_makes)]
sns.boxplot(x='make', y='sellingprice', data=df_top_makes)
plt.xticks(rotation=45)
plt.title('Distribution of Average Selling Prices by Brand(Make) (Top 10)')
plt.tight_layout()
plt.show()
```

```python
# BODY
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(25, 17))
plt.figure(figsize=(26, 25))

ax1.set_title('Number of cars by body')
sns.countplot(data['body'], order = data['body'].value_counts().index, ax=ax1)

ax2.set_title('Selling price by body')
sns.boxplot(x='sellingprice', y='body', data=data, order=data['body'].value_counts().index, ax=ax2)
```

```python
#average price by body type
avg_price_body = data.groupby('body')['sellingprice'].agg(['mean', 'count']).round(2)
```

```python
# AVG Selling Price by Body Type
plt.figure(figsize= (10,12))
avg_price_body['mean'].plot(kind='barh', color='skyblue')
plt.title('Average Selling Price by Body Type')
plt.xlabel('AVG Price')
plt.ylabel('Body Type')
plt.xticks(rotation=10)
plt.tight_layout()
print("\
Average Price by Body Type:")
print(avg_price_body)
plt.show()
```

## **Key Findings:**
Moderate positive relationship as years increase, prices tend to increase as well.

P-value of 0.000: The correlation is statistically significant (very unlikely to be due to chance).

* Newer cars and better condition are strongly associated with higher prices.

* Higher odometer readings (more miles) are associated with lower prices.
* Make and body type also play a significant role in determining price.
